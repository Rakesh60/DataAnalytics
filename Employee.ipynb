```python
x= df[['Salary']]
```

```python
y=df['Bonus']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mukesh=LinearRegression()
```

```python
#Training Models
mukesh.fit(x_train,y_train)
```

```python
y_pred=mukesh.predict(x_test)
```

```python
mukesh.predict([[61933]])
```

```python
#coefficient (c)
mukesh.intercept_
```

```python
#slope (m)
mukesh.coef_
```

```python
from sklearn.metrics import  r2_score,mean_squared_error
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
